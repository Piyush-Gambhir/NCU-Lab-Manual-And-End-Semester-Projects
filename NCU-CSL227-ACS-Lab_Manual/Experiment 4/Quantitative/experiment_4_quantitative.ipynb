### Leaf and Stem Plot


```python
# Create the stem and leaf plot

# marks obtained by students in an examination
marks = [15, 18, 19, 27, 31, 42, 44, 56, 68, 71, 72, 83, 85, 95]
# corresponding stems
stems = [1, 1, 1, 2, 3, 4, 4, 5, 6, 7, 7, 8, 8, 9]

# set the x axis and y axis limits
pylab.xlim([0, 10])
pylab.ylim([0, 100])

# Provide a title for the stem plot
plt.title('Stem Plot(Stem and Leaf Plot)', fontsize=16)

# Give x and y label for the stem plot
plt.xlabel('Range/Stems', fontsize=12)
plt.ylabel('Marks obtained/Leafs', fontsize=12)

# plot the stem plot using matplotlib
markerline, stemlines, baseline = plt.stem(
    stems, marks, '--')

plt.setp(markerline, color='red')
plt.setp(stemlines, color='red')

# display the stem plot
plt.show()
```

### Histogram: For Univariate Variable


```python
# Plotting histogram for age
plt.hist(data['age'], bins = 8, color='red', alpha=0.5)
plt.title('Histogram for Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(range(0, 100, 10))
plt.yticks(range(0, 300, 50))
plt.legend(['Age'])
plt.show()
```

### Ogive /Cumulative Frequency Polygon Chart for Univariate Quantitative data

```python

# we can use the numpy.histogram function to automatically find the classes and the class frequencies.
# Then we can use matplotlib to actually create the ogive:

# obtain histogram values with 10 bins
values, base = np.histogram(data['age'], bins=10)

# find the cumulative sums
cumulative = np.cumsum(values)

# plot the ogive
plt.plot(base[:-1], cumulative, 'ro-')
plt.xlabel('Age')
plt.ylabel('Cumulative Frequency')
plt.title('Ogive for Age')
plt.show()
```

### Scatter Plot


```python
# Plotting Scatter Plot
plt.scatter(data['age'], data['chol'], color='red', alpha = 0.1)
plt.title('Scatter Plot for Age and Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend(['Age', 'Weight'])
plt.show()
```
