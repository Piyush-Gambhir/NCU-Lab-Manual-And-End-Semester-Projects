```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data
df = pd.read_csv('Project_Dataset_SF Salaries.csv')
df
```

```python
# Check the head of the DataFrame.
df.head()
```

```python
# Use the .info() method to find out how many entries there are.
df.info()
```

```python
# Print Number of unique values in each column
for col in df.columns:
    print(col, df[col].nunique())
    
```

```python
# What is the average BasePay
df['BasePay'].mean()
```

```python
# What is the highest amount of OvertimePay in the dataset
df['OvertimePay'].max()
```

```python
# Graph the top 5 jobs with the most BasePay
df.groupby('JobTitle')['BasePay'].mean().sort_values(ascending=False).head(5).plot(kind='bar')
```
