8. Using loc and ilco to Acces Dataframe Rows and Columns

```python
# accessing rows and columns in a df using iloc and loc

# create a dataframe from a dictionary
df = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})

# access a row using iloc
print("Accessing row using iloc")
print(df.iloc[0].to_markdown())

print("\n")

# access a column using iloc
print("Accessing column using iloc")
print(df.iloc[:, 0].to_markdown())

print("\n")

# access a row using loc
print("Accessing row using loc")
print(df.loc[0].to_markdown())

print("\n")

# access a column using loc
print("Accessing column using loc")
print(df.loc[:, 'a'].to_markdown())
```
