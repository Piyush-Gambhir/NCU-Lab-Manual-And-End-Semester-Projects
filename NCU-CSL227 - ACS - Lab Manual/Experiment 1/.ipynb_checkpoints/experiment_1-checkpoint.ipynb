### Introduction to statistics using Python.


```python
# Importitng the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Importing the dataset
dataset = pd.read_csv('diabetes_dataset.csv')
```

```python
# Checking the top 5 rows of the dataset
print("Top 5 rows of the dataset: ")
print(dataset.head().to_markdown())
```

```python
# Replacing 0 and 1 in the Outcome column with Non-Diabetic and Diabetic respectively
dataset['Outcome'] = dataset['Outcome'].replace(0, 'Non-Diabetic')
dataset['Outcome'] = dataset['Outcome'].replace(1, 'Diabetic')
# Checking the top 5 rows of the dataset
print("Top 5 rows of the dataset: ")
print(dataset.head().to_markdown())
```

```python
# Checking the information of the dataset
print("Information of the dataset:")
print(dataset.info())
```

```python
# Checking the statistical description of the dataset
print("Statistical description of the dataset:")
print(dataset.describe().to_markdown())
```

```python
# Checking the number of null values in the dataset
null_values = dataset.isnull().sum()
print(null_values)
```

```python
# Checking the number of unique values in the dataset
unique_values = dataset.nunique()
print(unique_values)
```

```python
# Checking the number of Diabetic and Non-Diabetic patients
print("Number of Diabetic and Non-Diabetic patients:")
print(dataset['Outcome'].value_counts())
```

```python
# Plotting Histograms for all the features in the dataset to check the distribution of the data
for i in dataset.columns:
    plt.hist(dataset[i])
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.title("Histogram of " + i)
    plt.show()
```

```python
# Pairplot of the dataset
sns.pairplot(dataset, hue='Outcome')
plt.suptitle("Pairplot of all variables colored by Outcome", y=1.05, fontsize=16)
plt.show()
```

```python
# Boxplot of the dataset
fig, ax = plt.subplots(figsize=(10, 10))
sns.boxplot(data=dataset)
plt.title("Boxplot of all variables", fontsize=16)
plt.show()
```
