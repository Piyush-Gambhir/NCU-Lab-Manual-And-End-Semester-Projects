## Experiment 4 (Qualitative)
### Data visualization for qualitative data.

```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read the data
housing = pd.read_csv('housing_dataset.csv')
```

```python
# Printing the First 5 Rows of the Data
print(housing.head().to_markdown())
```

```python
# Describing the Data
print(housing.describe().to_markdown())
```

```python
# Information about the Data
print(housing.info())
```

```python
# Checking for Missing Values
print(housing.isnull().sum())
```

```python
# Plotting Histograms for each Numerical Feature
housing.hist(bins=50, figsize=(20, 15), color='red',
             alpha=0.5, edgecolor='black')
plt.show()
```

```python
# Plotting Pie Chart for the ocean-proximity feature
housing['ocean_proximity'].value_counts().plot(kind='pie', figsize=(20, 8), ylabel=" ", autopct='%1.2f%%', legend=True, fontsize=12, cmap='cool', title='Ocean Proximity')
plt.legend(loc='upper right')
plt.show()
```

```python
# Plotting Bar Plot   
housing['ocean_proximity'].value_counts().plot(kind='bar', figsize=(10, 10), ylabel="Frequency", xlabel="Ocean Proximity",legend=True, fontsize=12, cmap='cool', title='Ocean Proximity')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.gca().xaxis.labelpad = 30
plt.gca().yaxis.labelpad = 30
plt.show()
```

```python
# Plotting Box Plot for the Target Variable
housing['median_house_value'].plot(kind='box', figsize=(20, 10), ylabel=" ", legend=True, fontsize=12, cmap='cool', title='Median House Value')
```
